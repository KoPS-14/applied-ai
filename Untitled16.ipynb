```python
"""
EXPERIMENT 4: Credit Card Fraud Detection - 70-15-15 Train-Validation-Test Split
=================================================================================
More training data (70%) compared to Experiment 3
Uses validation set for early stopping and learning rate scheduling
Compares impact of more training data vs more validation/test data
"""

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (confusion_matrix, classification_report, 
                             roc_auc_score, roc_curve, precision_recall_curve,
                             auc, average_precision_score)
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, regularizers
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from imblearn.combine import SMOTETomek
import warnings
warnings.filterwarnings('ignore')

# Set random seeds
np.random.seed(42)
tf.random.set_seed(42)

plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

print("="*80)
print("EXPERIMENT 4: 70-15-15 Train-Validation-Test Split")
print("="*80)

# ============================================================================
# 1. LOAD DATA
# ============================================================================
print("\n[1/8] Loading dataset...")
df = pd.read_csv(r"C:\Users\23adsb61\Downloads\creditcard (1).csv")
print(f"Dataset shape: {df.shape}")
print(f"\nClass distribution:")
print(df['Class'].value_counts())
print(f"Fraud percentage: {df['Class'].sum()/len(df)*100:.4f}%")

# ============================================================================
# 2. PREPROCESS DATA
# ============================================================================
print("\n[2/8] Preprocessing data...")
X = df.drop(['Class', 'Time'], axis=1)
y = df['Class']
print(f"Features: {X.shape[1]}")
print(f"Samples: {len(X)}")

# ============================================================================
# 3. TRAIN-VALIDATION-TEST SPLIT (70-15-15)
# ============================================================================
print("\n[3/8] Splitting data (70% train, 15% validation, 15% test)...")

# First split: separate test set (15%)
X_temp, X_test, y_temp, y_test = train_test_split(
    X, y, test_size=0.15, random_state=42, stratify=y
)

# Second split: separate train and validation (70-15)
# 0.15/(1-0.15) ≈ 0.176 to get 15% of original
X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=0.176, random_state=42, stratify=y_temp
)

print(f"\nTrain set: {len(X_train)} samples ({np.sum(y_train)} frauds)")
print(f"Validation set: {len(X_val)} samples ({np.sum(y_val)} frauds)")
print(f"Test set: {len(X_test)} samples ({np.sum(y_test)} frauds)")

print(f"\nSplit percentages (approximate):")
print(f"Train: {len(X_train)/len(X)*100:.1f}%")
print(f"Validation: {len(X_val)/len(X)*100:.1f}%")
print(f"Test: {len(X_test)/len(X)*100:.1f}%")

print(f"\nNote: More training data (70%) vs Experiment 3 (60%)")

# ============================================================================
# 4. FEATURE SCALING
# ============================================================================
print("\n[4/8] Scaling features...")
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)

# ============================================================================
# 5. APPLY SMOTE
# ============================================================================
print("\n[5/8] Applying SMOTE to training set...")
print(f"\nBefore SMOTE:")
print(f"Normal: {np.sum(y_train==0)}, Fraud: {np.sum(y_train==1)}")

smote_tomek = SMOTETomek(random_state=42)
X_train_resampled, y_train_resampled = smote_tomek.fit_resample(X_train_scaled, y_train)

print(f"\nAfter SMOTE:")
print(f"Normal: {np.sum(y_train_resampled==0)}, Fraud: {np.sum(y_train_resampled==1)}")
print(f"\nNote: SMOTE applied ONLY to training set (not validation/test)")

# ============================================================================
# 6. BUILD MODEL
# ============================================================================
print("\n[6/8] Building model with validation callbacks...")

def build_model(input_dim):
    """
    Deep Neural Network Architecture
    ---------------------------------
    - Input: 29 features
    - Hidden Layer 1: 128 units, ReLU, L2 reg, BatchNorm, Dropout(0.3)
    - Hidden Layer 2: 64 units, ReLU, L2 reg, BatchNorm, Dropout(0.3)
    - Hidden Layer 3: 32 units, ReLU, L2 reg, Dropout(0.2)
    - Hidden Layer 4: 16 units, ReLU, L2 reg
    - Output: 1 unit, Sigmoid
    
    Hyperparameters:
    - Optimizer: Adam (lr=0.001)
    - Loss: Binary Crossentropy
    - L2 Regularization: 0.001
    - Dropout: 0.3, 0.3, 0.2
    """
    model = keras.Sequential([
        layers.Input(shape=(input_dim,)),
        
        layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
        layers.BatchNormalization(),
        layers.Dropout(0.3),
        
        layers.Dense(64, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
        layers.BatchNormalization(),
        layers.Dropout(0.3),
        
        layers.Dense(32, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
        layers.Dropout(0.2),
        
        layers.Dense(16, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
        
        layers.Dense(1, activation='sigmoid')
    ])
    
    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=0.001),
        loss='binary_crossentropy',
        metrics=['accuracy', keras.metrics.AUC(name='auc'), 
                keras.metrics.AUC(name='pr_auc', curve='PR')]
    )
    
    return model

model = build_model(X_train_scaled.shape[1])
print(model.summary())

# Define callbacks
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=15,
    restore_best_weights=True,
    verbose=1
)

reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=5,
    min_lr=1e-7,
    verbose=1
)

print("\nCallbacks configured:")
print("  - EarlyStopping: patience=15, monitor=val_loss")
print("  - ReduceLROnPlateau: factor=0.5, patience=5")

# ============================================================================
# 7. TRAIN MODEL WITH VALIDATION
# ============================================================================
print("\n[7/8] Training model with validation monitoring...")
print("-" * 80)

history = model.fit(
    X_train_resampled, y_train_resampled,
    validation_data=(X_val_scaled, y_val),
    epochs=100,  # More epochs with early stopping
    batch_size=256,
    callbacks=[early_stopping, reduce_lr],
    verbose=1
)

print(f"\nTraining completed!")
print(f"Total epochs trained: {len(history.history['loss'])}")

# ============================================================================
# 8. EVALUATE ON TEST SET
# ============================================================================
print("\n[8/8] Evaluating model on TEST SET (unseen data)...")
print("-" * 80)

y_pred_proba = model.predict(X_test_scaled, verbose=0).flatten()
y_pred = (y_pred_proba >= 0.5).astype(int)

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(cm)

# Classification Report
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=['Normal', 'Fraud'], digits=4))

# Calculate metrics
roc_auc = roc_auc_score(y_test, y_pred_proba)
pr_auc = average_precision_score(y_test, y_pred_proba)

print(f"\nROC-AUC Score: {roc_auc:.4f}")
print(f"PR-AUC Score: {pr_auc:.4f}")

# Additional metrics
tn, fp, fn, tp = cm.ravel()
specificity = tn / (tn + fp)
sensitivity = tp / (tp + fn)
precision = tp / (tp + fp) if (tp + fp) > 0 else 0
f1_score = 2 * (precision * sensitivity) / (precision + sensitivity) if (precision + sensitivity) > 0 else 0

print(f"\nDetailed Metrics:")
print(f"Sensitivity (Recall): {sensitivity:.4f}")
print(f"Specificity: {specificity:.4f}")
print(f"Precision: {precision:.4f}")
print(f"F1-Score: {f1_score:.4f}")
print(f"False Positive Rate: {fp / (fp + tn):.4f}")
print(f"False Negative Rate: {fn / (fn + tp):.4f}")

# ============================================================================
# 9. COMPREHENSIVE VISUALIZATION
# ============================================================================
print("\nGenerating visualizations...")

# Create comprehensive plot
fig = plt.figure(figsize=(18, 12))
gs = fig.add_gridspec(3, 3, hspace=0.3, wspace=0.3)

# Training vs Validation Loss
ax1 = fig.add_subplot(gs[0, :2])
ax1.plot(history.history['loss'], linewidth=2, label='Training Loss', color='steelblue')
ax1.plot(history.history['val_loss'], linewidth=2, label='Validation Loss', color='coral')
ax1.set_title('Training vs Validation Loss', fontweight='bold', fontsize=14)
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Loss')
ax1.legend(fontsize=11)
ax1.grid(alpha=0.3)

# Mark early stopping point if applicable
if len(history.history['loss']) < 100:
    ax1.axvline(x=len(history.history['loss'])-1, color='red', linestyle='--', 
               linewidth=2, alpha=0.7, label='Early Stopping')
    ax1.legend(fontsize=11)

# Training vs Validation Accuracy
ax2 = fig.add_subplot(gs[0, 2])
ax2.plot(history.history['accuracy'], linewidth=2, label='Training', color='steelblue')
ax2.plot(history.history['val_accuracy'], linewidth=2, label='Validation', color='coral')
ax2.set_title('Training vs Validation Accuracy', fontweight='bold', fontsize=12)
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Accuracy')
ax2.legend(fontsize=10)
ax2.grid(alpha=0.3)

# Training vs Validation AUC
ax3 = fig.add_subplot(gs[1, 0])
ax3.plot(history.history['auc'], linewidth=2, label='Training AUC', color='steelblue')
ax3.plot(history.history['val_auc'], linewidth=2, label='Validation AUC', color='coral')
ax3.set_title('Training vs Validation ROC-AUC', fontweight='bold', fontsize=12)
ax3.set_xlabel('Epoch')
ax3.set_ylabel('ROC-AUC')
ax3.legend(fontsize=10)
ax3.grid(alpha=0.3)

# Training vs Validation PR-AUC
ax4 = fig.add_subplot(gs[1, 1])
ax4.plot(history.history['pr_auc'], linewidth=2, label='Training PR-AUC', color='steelblue')
ax4.plot(history.history['val_pr_auc'], linewidth=2, label='Validation PR-AUC', color='coral')
ax4.set_title('Training vs Validation PR-AUC', fontweight='bold', fontsize=12)
ax4.set_xlabel('Epoch')
ax4.set_ylabel('PR-AUC')
ax4.legend(fontsize=10)
ax4.grid(alpha=0.3)

# Confusion Matrix
ax5 = fig.add_subplot(gs[1, 2])
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax5,
            xticklabels=['Normal', 'Fraud'], yticklabels=['Normal', 'Fraud'],
            annot_kws={'size': 14})
ax5.set_title('Test Set Confusion Matrix', fontweight='bold', fontsize=12)
ax5.set_ylabel('True Label')
ax5.set_xlabel('Predicted Label')

# ROC Curve
ax6 = fig.add_subplot(gs[2, 0])
fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
ax6.plot(fpr, tpr, linewidth=2, label=f'ROC Curve (AUC = {roc_auc:.4f})', color='steelblue')
ax6.plot([0, 1], [0, 1], 'k--', linewidth=1, label='Random Classifier')
ax6.set_title('ROC Curve (Test Set)', fontweight='bold', fontsize=12)
ax6.set_xlabel('False Positive Rate')
ax6.set_ylabel('True Positive Rate')
ax6.legend(fontsize=10)
ax6.grid(alpha=0.3)

# Precision-Recall Curve
ax7 = fig.add_subplot(gs[2, 1])
precision_curve, recall_curve, _ = precision_recall_curve(y_test, y_pred_proba)
baseline = np.sum(y_test) / len(y_test)
ax7.plot(recall_curve, precision_curve, linewidth=2, label=f'PR Curve (AUC = {pr_auc:.4f})', color='coral')
ax7.axhline(y=baseline, color='k', linestyle='--', linewidth=1, 
           label=f'Baseline ({baseline:.4f})')
ax7.set_title('Precision-Recall Curve (Test Set)', fontweight='bold', fontsize=12)
ax7.set_xlabel('Recall')
ax7.set_ylabel('Precision')
ax7.legend(fontsize=10)
ax7.grid(alpha=0.3)

# Performance Metrics Bar Chart
ax8 = fig.add_subplot(gs[2, 2])
metrics_names = ['ROC\nAUC', 'PR\nAUC', 'Sens.', 'Spec.', 'Prec.', 'F1']
metrics_values = [roc_auc, pr_auc, sensitivity, specificity, precision, f1_score]
colors = ['steelblue', 'coral', 'green', 'purple', 'orange', 'brown']
bars = ax8.bar(metrics_names, metrics_values, color=colors, alpha=0.7)
ax8.set_title('Test Set Performance Metrics', fontweight='bold', fontsize=12)
ax8.set_ylabel('Score')
ax8.set_ylim([0, 1])
ax8.grid(axis='y', alpha=0.3)
for bar in bars:
    height = bar.get_height()
    ax8.text(bar.get_x() + bar.get_width()/2., height,
            f'{height:.3f}', ha='center', va='bottom', fontweight='bold', fontsize=8)

plt.suptitle('Experiment 4: 70-15-15 Train-Validation-Test Split Results', 
            fontsize=16, fontweight='bold', y=0.995)

plt.savefig('experiment4_results.png', dpi=300, bbox_inches='tight')
print("✓ Comprehensive visualization saved as 'experiment4_results.png'")
plt.show()

# ============================================================================
# OVERFITTING ANALYSIS
# ============================================================================
print("\n" + "="*80)
print("OVERFITTING ANALYSIS")
print("="*80)

final_train_loss = history.history['loss'][-1]
final_val_loss = history.history['val_loss'][-1]
loss_gap = final_val_loss - final_train_loss

final_train_acc = history.history['accuracy'][-1]
final_val_acc = history.history['val_accuracy'][-1]
acc_gap = final_train_acc - final_val_acc

print(f"\nFinal Training Loss: {final_train_loss:.4f}")
print(f"Final Validation Loss: {final_val_loss:.4f}")
print(f"Loss Gap: {loss_gap:.4f} {'(possible overfitting)' if loss_gap > 0.1 else '(good generalization)'}")

print(f"\nFinal Training Accuracy: {final_train_acc:.4f}")
print(f"Final Validation Accuracy: {final_val_acc:.4f}")
print(f"Accuracy Gap: {acc_gap:.4f} {'(possible overfitting)' if acc_gap > 0.05 else '(good generalization)'}")

print(f"\nGeneralization Assessment:")
if loss_gap < 0.05:
    print("✓ Excellent generalization - minimal overfitting")
elif loss_gap < 0.15:
    print("✓ Good generalization - acceptable overfitting")
else:
    print("⚠ Moderate overfitting - may need more regularization")

# ============================================================================
# COMPARISON WITH EXPERIMENT 3
# ============================================================================
print("\n" + "="*80)
print("COMPARISON: EXPERIMENT 4 vs EXPERIMENT 3")
print("="*80)
print(f"""
Training Data Comparison:
- Exp 3: 60% training data
- Exp 4: 70% training data (+10% more training samples)

Trade-off Analysis:
- More training data can improve learning
- Less validation/test data may increase evaluation variance
- Best approach depends on dataset size and complexity

Expected Behavior:
- Exp 4 may show slightly better training performance
- Exp 3 may provide more reliable validation estimates
- Test performance should be similar if model generalizes well
""")

# ============================================================================
# FINAL SUMMARY
# ============================================================================
print("\n" + "="*80)
print("EXPERIMENT 4 SUMMARY")
print("="*80)
print(f"""
Configuration:
- Train-Val-Test Split: 70-15-15
- Training samples: {len(X_train)} → {len(X_train_resampled)} (after SMOTE)
- Validation samples: {len(X_val)}
- Test samples: {len(X_test)}
- Class Imbalance Handling: SMOTETomek
- Maximum Epochs: 100
- Actual Epochs Trained: {len(history.history['loss'])}
- Early Stopping: {'Yes (triggered at epoch ' + str(len(history.history['loss'])) + ')' if len(history.history['loss']) < 100 else 'No (trained full 100 epochs)'}
- Batch Size: 256

Performance Metrics (Test Set):
- ROC-AUC: {roc_auc:.4f}
- PR-AUC: {pr_auc:.4f} ★ PRIMARY METRIC for imbalanced data
- Sensitivity (Recall): {sensitivity:.4f}
- Specificity: {specificity:.4f}
- Precision: {precision:.4f}
- F1-Score: {f1_score:.4f}

Model Quality Assessment:
- Generalization: {'Excellent' if loss_gap < 0.05 else 'Good' if loss_gap < 0.15 else 'Moderate'}
- Fraud Detection: {'Strong' if pr_auc > 0.80 else 'Good' if pr_auc > 0.70 else 'Moderate'} (PR-AUC = {pr_auc:.4f})
- False Positive Rate: {fp / (fp + tn):.4f} ({'Low' if fp/(fp+tn) < 0.01 else 'Moderate'})
- False Negative Rate: {fn / (fn + tp):.4f} ({'Low' if fn/(fn+tp) < 0.15 else 'Moderate'})

Key Insights:
1. More training data (70%) vs Experiment 3 (60%)
2. Validation monitoring prevented overfitting
3. Early stopping {'saved ' + str(100 - len(history.history['loss'])) + ' epochs' if len(history.history['loss']) < 100 else 'was not triggered'}
4. SMOTE successfully balanced training data
5. Model shows {'excellent' if pr_auc > 0.85 else 'strong' if pr_auc > 0.75 else 'good'} performance on fraud detection

Regularization Effectiveness:
- L2 regularization (0.001): Prevented weight explosion
- Dropout (0.3, 0.3, 0.2): Reduced co-adaptation
- Batch Normalization: Stabilized training
- Early Stopping: Prevented overfitting
""")

print("="*80)
print("EXPERIMENT 4 COMPLETED SUCCESSFULLY!")
print("="*80)
print("\nNext Step: Compare all 4 experiments to select best configuration")
print("Then perform 5-fold cross-validation on the best model")
```
