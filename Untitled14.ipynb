```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import (confusion_matrix, classification_report, 
                             roc_auc_score, roc_curve, precision_recall_curve,
                             auc, average_precision_score)
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, regularizers
from imblearn.combine import SMOTETomek
import warnings
warnings.filterwarnings('ignore')

# Set random seeds
np.random.seed(42)
tf.random.set_seed(42)

plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

print("="*80)
print("EXPERIMENT 2: 70-30 Train-Test Split (No Validation)")
print("="*80)

# ============================================================================
# 1. LOAD DATA
# ============================================================================
print("\n[1/7] Loading dataset...")
df = pd.read_csv(r"C:\Users\23adsb61\Downloads\creditcard (1).csv")

print(f"Dataset shape: {df.shape}")
print(f"\nClass distribution:")
print(df['Class'].value_counts())
print(f"Fraud percentage: {df['Class'].sum()/len(df)*100:.4f}%")

# ============================================================================
# 2. PREPROCESS DATA
# ============================================================================
print("\n[2/7] Preprocessing data...")
X = df.drop(['Class', 'Time'], axis=1)
y = df['Class']
print(f"Features: {X.shape[1]}")
print(f"Samples: {len(X)}")

# ============================================================================
# 3. TRAIN-TEST SPLIT (70-30)
# ============================================================================
print("\n[3/7] Splitting data (70% train, 30% test)...")
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

print(f"Train set: {len(X_train)} samples ({np.sum(y_train)} frauds)")
print(f"Test set: {len(X_test)} samples ({np.sum(y_test)} frauds)")
print(f"\nNote: Larger test set (30%) provides more robust evaluation")

# ============================================================================
# 4. FEATURE SCALING
# ============================================================================
print("\n[4/7] Scaling features...")
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# ============================================================================
# 5. APPLY SMOTE
# ============================================================================
print("\n[5/7] Applying SMOTE for class imbalance...")
print(f"\nBefore SMOTE:")
print(f"Normal: {np.sum(y_train==0)}, Fraud: {np.sum(y_train==1)}")

smote_tomek = SMOTETomek(random_state=42)
X_train_resampled, y_train_resampled = smote_tomek.fit_resample(X_train_scaled, y_train)

print(f"\nAfter SMOTE:")
print(f"Normal: {np.sum(y_train_resampled==0)}, Fraud: {np.sum(y_train_resampled==1)}")

# ============================================================================
# 6. BUILD MODEL
# ============================================================================
print("\n[6/7] Building and training model...")

def build_model(input_dim):
    """
    Deep Neural Network Architecture
    ---------------------------------
    - Input: 29 features
    - Hidden Layer 1: 128 units, ReLU, L2 reg, BatchNorm, Dropout(0.3)
    - Hidden Layer 2: 64 units, ReLU, L2 reg, BatchNorm, Dropout(0.3)
    - Hidden Layer 3: 32 units, ReLU, L2 reg, Dropout(0.2)
    - Hidden Layer 4: 16 units, ReLU, L2 reg
    - Output: 1 unit, Sigmoid
    """
    model = keras.Sequential([
        layers.Input(shape=(input_dim,)),
        
        layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
        layers.BatchNormalization(),
        layers.Dropout(0.3),
        
        layers.Dense(64, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
        layers.BatchNormalization(),
        layers.Dropout(0.3),
        
        layers.Dense(32, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
        layers.Dropout(0.2),
        
        layers.Dense(16, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
        
        layers.Dense(1, activation='sigmoid')
    ])
    
    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=0.001),
        loss='binary_crossentropy',
        metrics=['accuracy', keras.metrics.AUC(name='auc'), 
                keras.metrics.AUC(name='pr_auc', curve='PR')]
    )
    
    return model

model = build_model(X_train_scaled.shape[1])
print(model.summary())

# Train model
print("\nTraining model...")
history = model.fit(
    X_train_resampled, y_train_resampled,
    epochs=50,
    batch_size=256,
    verbose=1
)

# ============================================================================
# 7. EVALUATE MODEL
# ============================================================================
print("\n[7/7] Evaluating model on test set...")
print("-" * 80)

y_pred_proba = model.predict(X_test_scaled, verbose=0).flatten()
y_pred = (y_pred_proba >= 0.5).astype(int)

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(cm)

# Classification Report
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=['Normal', 'Fraud']))

# Calculate metrics
roc_auc = roc_auc_score(y_test, y_pred_proba)
pr_auc = average_precision_score(y_test, y_pred_proba)

print(f"\nROC-AUC Score: {roc_auc:.4f}")
print(f"PR-AUC Score: {pr_auc:.4f}")

# Additional metrics
tn, fp, fn, tp = cm.ravel()
specificity = tn / (tn + fp)
sensitivity = tp / (tp + fn)

print(f"\nAdditional Metrics:")
print(f"Sensitivity (Recall): {sensitivity:.4f}")
print(f"Specificity: {specificity:.4f}")
print(f"False Positive Rate: {fp / (fp + tn):.4f}")
print(f"False Negative Rate: {fn / (fn + tp):.4f}")

# ============================================================================
# 8. VISUALIZATION
# ============================================================================
print("\nGenerating visualizations...")

# Create comprehensive plot
fig = plt.figure(figsize=(16, 10))
gs = fig.add_gridspec(3, 2, hspace=0.3, wspace=0.3)

# Training Loss
ax1 = fig.add_subplot(gs[0, 0])
ax1.plot(history.history['loss'], linewidth=2, color='steelblue')
ax1.set_title('Training Loss Curve', fontweight='bold', fontsize=12)
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Loss')
ax1.grid(alpha=0.3)

# Training Accuracy
ax2 = fig.add_subplot(gs[0, 1])
ax2.plot(history.history['accuracy'], linewidth=2, color='coral')
ax2.set_title('Training Accuracy Curve', fontweight='bold', fontsize=12)
ax2.set_xlabel('Epoch')
ax2.set_ylabel('Accuracy')
ax2.grid(alpha=0.3)

# Confusion Matrix
ax3 = fig.add_subplot(gs[1, 0])
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax3,
            xticklabels=['Normal', 'Fraud'], yticklabels=['Normal', 'Fraud'],
            annot_kws={'size': 14})
ax3.set_title('Confusion Matrix', fontweight='bold', fontsize=12)
ax3.set_ylabel('True Label')
ax3.set_xlabel('Predicted Label')

# Metrics Bar Chart
ax4 = fig.add_subplot(gs[1, 1])
metrics_names = ['ROC-AUC', 'PR-AUC', 'Sensitivity', 'Specificity']
metrics_values = [roc_auc, pr_auc, sensitivity, specificity]
bars = ax4.bar(metrics_names, metrics_values, color=['steelblue', 'coral', 'green', 'purple'], alpha=0.7)
ax4.set_title('Performance Metrics', fontweight='bold', fontsize=12)
ax4.set_ylabel('Score')
ax4.set_ylim([0, 1])
ax4.grid(axis='y', alpha=0.3)
for bar in bars:
    height = bar.get_height()
    ax4.text(bar.get_x() + bar.get_width()/2., height,
            f'{height:.4f}', ha='center', va='bottom', fontweight='bold')

# ROC Curve
ax5 = fig.add_subplot(gs[2, 0])
fpr, tpr, _ = roc_curve(y_test, y_pred_proba)
ax5.plot(fpr, tpr, linewidth=2, label=f'ROC Curve (AUC = {roc_auc:.4f})', color='steelblue')
ax5.plot([0, 1], [0, 1], 'k--', linewidth=1, label='Random Classifier')
ax5.set_title('ROC Curve', fontweight='bold', fontsize=12)
ax5.set_xlabel('False Positive Rate')
ax5.set_ylabel('True Positive Rate')
ax5.legend()
ax5.grid(alpha=0.3)

# Precision-Recall Curve
ax6 = fig.add_subplot(gs[2, 1])
precision, recall, _ = precision_recall_curve(y_test, y_pred_proba)
baseline = np.sum(y_test) / len(y_test)
ax6.plot(recall, precision, linewidth=2, label=f'PR Curve (AUC = {pr_auc:.4f})', color='coral')
ax6.axhline(y=baseline, color='k', linestyle='--', linewidth=1, 
           label=f'Baseline ({baseline:.4f})')
ax6.set_title('Precision-Recall Curve', fontweight='bold', fontsize=12)
ax6.set_xlabel('Recall')
ax6.set_ylabel('Precision')
ax6.legend()
ax6.grid(alpha=0.3)

plt.suptitle('Experiment 2: 70-30 Train-Test Split Results', 
            fontsize=16, fontweight='bold', y=0.995)

plt.savefig('experiment2_results.png', dpi=300, bbox_inches='tight')
print("✓ Visualization saved as 'experiment2_results.png'")
plt.show()

# ============================================================================
# FINAL SUMMARY
# ============================================================================
print("\n" + "="*80)
print("EXPERIMENT 2 SUMMARY")
print("="*80)
print(f"""
Configuration:
- Train-Test Split: 70-30 (no validation set)
- Training samples: {len(X_train)} → {len(X_train_resampled)} (after SMOTE)
- Test samples: {len(X_test)}
- Class Imbalance Handling: SMOTETomek
- Training Epochs: 50
- Batch Size: 256

Performance Metrics:
- ROC-AUC: {roc_auc:.4f}
- PR-AUC: {pr_auc:.4f} (PRIMARY METRIC for imbalanced data)
- Sensitivity (Recall): {sensitivity:.4f}
- Specificity: {specificity:.4f}

Key Observations:
1. Larger test set (30%) provides more robust evaluation
2. SMOTE successfully balanced the training data
3. Model shows {('good' if pr_auc > 0.75 else 'moderate')} performance on minority class
4. {'Low' if fp < 100 else 'Moderate'} false positive rate important for operational costs
""")

print("="*80)
print("EXPERIMENT 2 COMPLETED SUCCESSFULLY!")
print("="*80)
```
