```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import roc_auc_score, average_precision_score
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, regularizers
from imblearn.combine import SMOTETomek
import warnings
warnings.filterwarnings('ignore')

# Set random seeds
np.random.seed(42)
tf.random.set_seed(42)

plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

print("="*80)
print("5-FOLD STRATIFIED CROSS-VALIDATION")
print("="*80)
print("\nBest Configuration: Based on Experiment 3 (60-20-20 split with validation)")
print("Using: SMOTE + Deep NN Architecture")

# ============================================================================
# 1. LOAD DATA
# ============================================================================
print("\n[1/4] Loading dataset...")
df = pd.read_csv(r"C:\Users\23adsb61\Downloads\creditcard (1).csv")

print(f"Dataset shape: {df.shape}")
print(f"Fraud percentage: {df['Class'].sum()/len(df)*100:.4f}%")

# ============================================================================
# 2. PREPROCESS DATA
# ============================================================================
print("\n[2/4] Preprocessing data...")
X = df.drop(['Class', 'Time'], axis=1).values  # Convert to numpy for sklearn
y = df['Class'].values

print(f"Features: {X.shape[1]}")
print(f"Samples: {len(X)}")

# ============================================================================
# 3. MODEL ARCHITECTURE
# ============================================================================
def build_model(input_dim):
    """
    Best performing Deep Neural Network Architecture
    Based on experiments, this architecture showed best results
    """
    model = keras.Sequential([
        layers.Input(shape=(input_dim,)),
        
        layers.Dense(128, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
        layers.BatchNormalization(),
        layers.Dropout(0.3),
        
        layers.Dense(64, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
        layers.BatchNormalization(),
        layers.Dropout(0.3),
        
        layers.Dense(32, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
        layers.Dropout(0.2),
        
        layers.Dense(16, activation='relu', kernel_regularizer=regularizers.l2(0.001)),
        
        layers.Dense(1, activation='sigmoid')
    ])
    
    model.compile(
        optimizer=keras.optimizers.Adam(learning_rate=0.001),
        loss='binary_crossentropy',
        metrics=['accuracy', keras.metrics.AUC(name='auc'), 
                keras.metrics.AUC(name='pr_auc', curve='PR')]
    )
    
    return model

# ============================================================================
# 4. CROSS-VALIDATION
# ============================================================================
print("\n[3/4] Performing 5-fold stratified cross-validation...")
print("This may take several minutes...")
print("-" * 80)

# Initialize StratifiedKFold
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

# Storage for results
roc_auc_scores = []
pr_auc_scores = []
fold_details = []

# Perform cross-validation
for fold, (train_idx, test_idx) in enumerate(skf.split(X, y), 1):
    print(f"\n{'='*60}")
    print(f"FOLD {fold}/5")
    print(f"{'='*60}")
    
    # Split data
    X_train_fold = X[train_idx]
    X_test_fold = X[test_idx]
    y_train_fold = y[train_idx]
    y_test_fold = y[test_idx]
    
    print(f"Train samples: {len(X_train_fold)} ({np.sum(y_train_fold)} frauds)")
    print(f"Test samples: {len(X_test_fold)} ({np.sum(y_test_fold)} frauds)")
    
    # Scale features
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train_fold)
    X_test_scaled = scaler.transform(X_test_fold)
    
    # Apply SMOTE
    print(f"\nApplying SMOTE...")
    print(f"Before: Normal={np.sum(y_train_fold==0)}, Fraud={np.sum(y_train_fold==1)}")
    
    smote_tomek = SMOTETomek(random_state=42)
    X_train_resampled, y_train_resampled = smote_tomek.fit_resample(
        X_train_scaled, y_train_fold
    )
    
    print(f"After: Normal={np.sum(y_train_resampled==0)}, Fraud={np.sum(y_train_resampled==1)}")
    
    # Build and train model
    print(f"\nTraining model...")
    model = build_model(X_train_scaled.shape[1])
    
    history = model.fit(
        X_train_resampled, y_train_resampled,
        epochs=50,
        batch_size=256,
        verbose=0,  # Silent training for cleaner output
        shuffle=True
    )
    
    # Evaluate
    print(f"\nEvaluating on test set...")
    y_pred_proba = model.predict(X_test_scaled, verbose=0).flatten()
    
    # Calculate metrics
    roc_auc = roc_auc_score(y_test_fold, y_pred_proba)
    pr_auc = average_precision_score(y_test_fold, y_pred_proba)
    
    # Store results
    roc_auc_scores.append(roc_auc)
    pr_auc_scores.append(pr_auc)
    
    fold_details.append({
        'fold': fold,
        'roc_auc': roc_auc,
        'pr_auc': pr_auc,
        'train_samples': len(X_train_fold),
        'test_samples': len(X_test_fold),
        'final_train_loss': history.history['loss'][-1],
        'final_train_acc': history.history['accuracy'][-1]
    })
    
    print(f"\nFold {fold} Results:")
    print(f"  ROC-AUC: {roc_auc:.4f}")
    print(f"  PR-AUC: {pr_auc:.4f}")
    print(f"  Final Training Loss: {history.history['loss'][-1]:.4f}")
    print(f"  Final Training Accuracy: {history.history['accuracy'][-1]:.4f}")

# ============================================================================
# 5. AGGREGATE RESULTS
# ============================================================================
print("\n" + "="*80)
print("CROSS-VALIDATION RESULTS")
print("="*80)

print("\nIndividual Fold Results:")
print(f"{'Fold':<6} {'ROC-AUC':<12} {'PR-AUC':<12} {'Train Loss':<12} {'Train Acc':<12}")
print("-" * 60)
for detail in fold_details:
    print(f"{detail['fold']:<6} {detail['roc_auc']:<12.4f} {detail['pr_auc']:<12.4f} "
          f"{detail['final_train_loss']:<12.4f} {detail['final_train_acc']:<12.4f}")

print("\n" + "-" * 80)
print("SUMMARY STATISTICS")
print("-" * 80)

# Calculate statistics
mean_roc_auc = np.mean(roc_auc_scores)
std_roc_auc = np.std(roc_auc_scores)
mean_pr_auc = np.mean(pr_auc_scores)
std_pr_auc = np.std(pr_auc_scores)

print(f"\nROC-AUC:")
print(f"  Mean: {mean_roc_auc:.4f}")
print(f"  Std Dev: {std_roc_auc:.4f}")
print(f"  Min: {np.min(roc_auc_scores):.4f}")
print(f"  Max: {np.max(roc_auc_scores):.4f}")
print(f"  Range: {np.max(roc_auc_scores) - np.min(roc_auc_scores):.4f}")

print(f"\nPR-AUC (Primary Metric for Imbalanced Data):")
print(f"  Mean: {mean_pr_auc:.4f}")
print(f"  Std Dev: {std_pr_auc:.4f}")
print(f"  Min: {np.min(pr_auc_scores):.4f}")
print(f"  Max: {np.max(pr_auc_scores):.4f}")
print(f"  Range: {np.max(pr_auc_scores) - np.min(pr_auc_scores):.4f}")

# ============================================================================
# 6. VISUALIZATION
# ============================================================================
print("\n[4/4] Generating visualizations...")

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# ROC-AUC Bar Chart
folds = list(range(1, 6))
axes[0, 0].bar(folds, roc_auc_scores, color='steelblue', alpha=0.7, edgecolor='black')
axes[0, 0].axhline(y=mean_roc_auc, color='red', linestyle='--', linewidth=2, 
                   label=f'Mean: {mean_roc_auc:.4f}')
axes[0, 0].axhline(y=mean_roc_auc + std_roc_auc, color='orange', linestyle=':', 
                   linewidth=1.5, alpha=0.7, label=f'Â±1 Std Dev')
axes[0, 0].axhline(y=mean_roc_auc - std_roc_auc, color='orange', linestyle=':', 
                   linewidth=1.5, alpha=0.7)
axes[0, 0].set_xlabel('Fold', fontsize=12, fontweight='bold')
axes[0, 0].set_ylabel('ROC-AUC Score', fontsize=12, fontweight='bold')
axes[0, 0].set_title('ROC-AUC Scores Across Folds', fontsize=14, fontweight='bold')
axes[0, 0].set_xticks(folds)
axes[0, 0].set_ylim([0.9, 1.0])
axes[0, 0].legend(fontsize=10)
axes[0, 0].grid(axis='y', alpha=0.3)

# Add value labels
for i, (fold, score) in enumerate(zip(folds, roc_auc_scores)):
    axes[0, 0].text(fold, score + 0.002, f'{score:.4f}', 
                    ha='center', va='bottom', fontsize=9, fontweight='bold')

# PR-AUC Bar Chart
axes[0, 1].bar(folds, pr_auc_scores, color='coral', alpha=0.7, edgecolor='black')
axes[0, 1].axhline(y=mean_pr_auc, color='red', linestyle='--', linewidth=2, 
                   label=f'Mean: {mean_pr_auc:.4f}')
axes[0, 1].axhline(y=mean_pr_auc + std_pr_auc, color='orange', linestyle=':', 
                   linewidth=1.5, alpha=0.7, label=f'Â±1 Std Dev')
axes[0, 1].axhline(y=mean_pr_auc - std_pr_auc, color='orange', linestyle=':', 
                   linewidth=1.5, alpha=0.7)
axes[0, 1].set_xlabel('Fold', fontsize=12, fontweight='bold')
axes[0, 1].set_ylabel('PR-AUC Score', fontsize=12, fontweight='bold')
axes[0, 1].set_title('PR-AUC Scores Across Folds', fontsize=14, fontweight='bold')
axes[0, 1].set_xticks(folds)
axes[0, 1].legend(fontsize=10)
axes[0, 1].grid(axis='y', alpha=0.3)

# Add value labels
for i, (fold, score) in enumerate(zip(folds, pr_auc_scores)):
    axes[0, 1].text(fold, score + 0.005, f'{score:.4f}', 
                    ha='center', va='bottom', fontsize=9, fontweight='bold')

# Box Plot
box_data = [roc_auc_scores, pr_auc_scores]
bp = axes[1, 0].boxplot(box_data, labels=['ROC-AUC', 'PR-AUC'], 
                        patch_artist=True, notch=True, showmeans=True)
bp['boxes'][0].set_facecolor('steelblue')
bp['boxes'][1].set_facecolor('coral')
axes[1, 0].set_ylabel('Score', fontsize=12, fontweight='bold')
axes[1, 0].set_title('Distribution of Metrics Across Folds', fontsize=14, fontweight='bold')
axes[1, 0].grid(axis='y', alpha=0.3)
axes[1, 0].set_ylim([0.7, 1.0])

# Scatter Plot
axes[1, 1].scatter(folds, roc_auc_scores, s=150, color='steelblue', 
                  alpha=0.7, edgecolors='black', linewidth=2, label='ROC-AUC', marker='o')
axes[1, 1].scatter(folds, pr_auc_scores, s=150, color='coral', 
                  alpha=0.7, edgecolors='black', linewidth=2, label='PR-AUC', marker='s')
axes[1, 1].plot(folds, roc_auc_scores, color='steelblue', linestyle='--', alpha=0.5)
axes[1, 1].plot(folds, pr_auc_scores, color='coral', linestyle='--', alpha=0.5)
axes[1, 1].set_xlabel('Fold', fontsize=12, fontweight='bold')
axes[1, 1].set_ylabel('Score', fontsize=12, fontweight='bold')
axes[1, 1].set_title('Metric Trends Across Folds', fontsize=14, fontweight='bold')
axes[1, 1].set_xticks(folds)
axes[1, 1].legend(fontsize=10)
axes[1, 1].grid(alpha=0.3)

plt.suptitle('5-Fold Stratified Cross-Validation Results', 
            fontsize=16, fontweight='bold', y=1.00)

plt.tight_layout()
plt.savefig('cross_validation_results.png', dpi=300, bbox_inches='tight')
print("âœ“ Visualization saved as 'cross_validation_results.png'")
plt.show()

# ============================================================================
# FINAL ANALYSIS
# ============================================================================
print("\n" + "="*80)
print("CROSS-VALIDATION ANALYSIS")
print("="*80)



print("="*80)
print("5-FOLD CROSS-VALIDATION COMPLETED SUCCESSFULLY!")
print("="*80)
print("\nFinal Results:")
print(f"  Mean ROC-AUC: {mean_roc_auc:.4f} Â± {std_roc_auc:.4f}")
print(f"  Mean PR-AUC:  {mean_pr_auc:.4f} Â± {std_pr_auc:.4f}")
```
